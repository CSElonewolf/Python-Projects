```python
import os 
import pandas as pd 
```

```python
cwd = os.getcwd()
DIR_NAM = os.path.dirname(os.getcwd())
BASE_DIR = os.path.join(DIR_NAM,'7. Build a Data Science Pipeline')
DATA_DIR = os.path.join(BASE_DIR,'data')
CACHE_DIR = os.path.join(BASE_DIR,'cache')
os.makedirs(CACHE_DIR,exist_ok=True)
```

```python
my_data = os.path.join(DATA_DIR,'2019.csv')
df = pd.read_csv(my_data)
df
```

```python
df['Year'] = 2019
```

```python

my_dataframe = []
csv_files = [x.replace('.csv','') for x in os.listdir(DATA_DIR) if x.endswith('.csv')]
csv_files.sort(reverse = True)
print(csv_files)
for filename in csv_files:
    year = filename
    csv_path = os.path.join(DATA_DIR,filename +'.csv')
    this_df = pd.read_csv(csv_path)
    this_df['filename'] = filename + '.csv'
    this_df['year'] = year
    my_dataframe.append(this_df)
```

```python
my_entire_dataframe = pd.concat(my_dataframe)
my_entire_dataframe.head()
```

```python
dataset = os.path.join(CACHE_DIR,'movies-box-office-dataset.csv')
my_entire_dataframe.to_csv(dataset,index= False)
```
