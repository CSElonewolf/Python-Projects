```python
import requests
from requests_html import HTML
from selenium import webdriver 
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options = options)
```

```python
url = "https://www.amazon.in/Lifelong-LLTM09-Motorized-HealthifyMe-Installation/dp/B07R4M527C/ref=sr_1_3?dchild=1&keywords=Treadmill&qid=1632209117&qsid=257-9065962-6350157&sr=8-3&sres=B07R4M527C%2CB07STW8MMW%2CB07T3H61JM%2CB07H5JBRD7%2CB07H3X7YZ8%2CB077P72P5W%2CB0846LQ4XW%2CB07DD2G71S%2CB08D797Q2F%2CB08HLSGPDH%2CB07NZPG1F6%2CB07SRTPV2M%2CB08R8BMKMM%2CB08VW2TZGC%2CB0886Y7ZLS%2CB09B3WD4DG&srpt=TREADMILL"
title_lookup="#productTitle"
price_lookup = "#priceblock_ourprice"
```

```python
driver.get(url)
body_el = driver.find_element_by_css_selector("body")
html_str = body_el.get_attribute("innerHTML")
print(html_str)
```

```python
html_obj = HTML(html=html_str)
html_obj.find(title_lookup, first=True)
```

```python
product_title = html_obj.find(title_lookup, first=True).text
product_price = html_obj.find(price_lookup, first=True).text
```
